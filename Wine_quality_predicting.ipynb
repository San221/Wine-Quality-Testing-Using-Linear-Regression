```python
features=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
x=wine_data[features]
y=wine_data['quality']
sns.pairplot(wine_data,x_vars=features, y_vars='quality',kind='reg',size=7, aspect=0.5)
```

## Training the data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0) 
```

```python
regressor = LinearRegression()  
regressor.fit(x_train, y_train) 
```

```python
 #intercept and coefficient
print(regressor.intercept_)
print(regressor.coef_)
```

## Predicting 

```python
y_pred = regressor.predict(x_test)
```

```python
Parameters_obtained = [[7.5,0.8,0.02,2.0,0.075,12.0,35.0,0.9977,3.51,0.58,11]]
my_pred = regressor.predict(Parameters_obtained)
print("Parameters_obtained = {}".format(Parameters_obtained))
print("Predicted quality = {}".format(my_pred[0]))
```

## Evaluation

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
